```python
import numpy as np
import matplotlib.pyplot as plt

keys_sorted = sorted(original_precision.keys())
original_values_sorted = [original_precision[k] for k in keys_sorted]
final_values_sorted = [final_precision[k] for k in keys_sorted]

# Converting to numpy arrays
x = np.array(keys_sorted)
original_y = np.array(original_values_sorted)
final_y = np.array(final_values_sorted)

# Plotting
plt.figure(figsize=(10, 6))
scatter1 = plt.scatter(x, original_y, color='blue')
line1, = plt.plot(x, original_y, color='blue', linestyle='--', label='Original Precision')
scatter2 = plt.scatter(x, final_y, color='red')
line2, = plt.plot(x, final_y, color='red', label='Final Precision')

# Setting plot details
plt.xlabel('Precision@', fontsize=15)
plt.ylabel('Results', fontsize=15)
plt.xticks(x)
plt.ylim(0.3, 0.8)
plt.title('Comparison of Original and Final Precision@k')
plt.legend(handles=[line1, line2], loc='upper right')
plt.grid(True, axis='y')
plt.savefig('Comparison_Precision.pdf')
plt.show()
```
